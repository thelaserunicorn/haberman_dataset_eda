```python
sns.FacetGrid(haberman,hue='Survival_Status', height=10)\
    .map(sns.distplot,'Axillary_Nodes')\
    .add_legend()
```

The above code snippet is used to visualize PDF of Axillary Nodes and it is observed that people with less than 5 Axillary Nodes usually survives longer although the data still overlaps a lot this is the best data to choose among all other PDF from the rest of the features that are given.

```python
haberman_Long_Survive = haberman.loc[haberman['Survival_Status'] == 1];
haberman_Short_Survive = haberman.loc[haberman['Survival_Status'] == 2];


counts, bin_edges = np.histogram(haberman_Long_Survive['Axillary_Nodes'], bins=10, 
 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(haberman_Short_Survive['Axillary_Nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.show();
```

From above CDF we can observe that orange line shows there is a 85% chance of long survival if number of axillary nodes detected are < 5. Also you can see as number of axillary nodes increases survival chances also reduces means it is clearly observed that 80%  85% of people have good chances of survival if they have less no of auxillary nodes detected and as nodes increases the survival status also decreases as a result 100% of people have less chances of survival if nodes increases >40

We can observe in above combined CDF for Long survival observation is same but in Short survival nearly 55% of people who have nodes less than 5 and there are nearly 100% of people in short survival if nodes are > 40

```python
print('Medians:')
print(np.median(haberman_Long_Survive['Axillary_Nodes']))

print(np.median(haberman_Short_Survive['Axillary_Nodes']))

print('\nQuantiles:')
print(np.percentile(haberman_Long_Survive['Axillary_Nodes'],np.arange(0,100,25)))
print(np.percentile(haberman_Short_Survive['Axillary_Nodes'],np.arange(0,100,25)))
```

```python
print('\n90th percentile:')
print(np.percentile(haberman_Long_Survive['Axillary_Nodes'],90))
print(np.percentile(haberman_Short_Survive['Axillary_Nodes'],90))
```

# Observation:

From above observation it is clear that average axillary nodes in long survival is 0 and for short survival it is 4. Patients who have average 4 auxillary nodes have short survival status.
Quantiles shows that nearly 50% of axillary nodes are 0 in long survival and 75% of patients have nodes less than 3 that is 25% patients are having nodes more than 3.
Similarly, In short survival 75% of patients have minimum 11 nodes detected.
90% of patients with long survival rate has less than 8nodes unlike short survival rate patients who generally had 20 nodes.

```python
sns.violinplot(x='Survival_Status', y='Axillary_Nodes',data=haberman)
plt.legend
plt.show()
```

Above violin plot (blue) concludes that for long survival rate a person should have nodes near 0

```python
sns.jointplot(x='Age',y='Axillary_Nodes',data=haberman_Long_Survive,kind='kde',height=8)
plt.grid()
plt.show()
```

Observation: Above is the 2D density plot for long survival using feature age and axillary nodes, it is observed the density of point for long survival is more for people with axillary nodes from 03 and usually people age from 40-65 is effected by breast cancer. The dark area have major density which is hill top in 3D and density is getting low as graph get lighter. Each shade represent 1 contour plot.

```python
sns.jointplot(x='Age',y='Axillary_Nodes',data=haberman_Short_Survive,kind='kde',height=8)
plt.grid()
plt.show()
```

Observation: people with short survival status has age within 50 with lymph nodes averaging at about 5 and is effected by breast cancer.
